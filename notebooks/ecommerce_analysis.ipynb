# ==============================
# 1. Imports
# ==============================

```python
import pandas as pd
import numpy as mp
import matplotlib.pyplot as plt
```

# =====================================
# 2. Load Dataset
# =====================================

```python
df = pd.read_csv("../data/Global_SuperStore.csv")
```

# =====================================
# 3. Data Preview
# =====================================

```python
df.head()
df.info()
df.describe()
df.isnull().sum()
```

# =====================================
# 4. KPI Analysis
# =====================================

```python
total_sales = df["Sales"].sum()

total_profit = df["Profit"].sum()

total_orders = df["Order ID"].nunique()

total_customers = df["Customer ID"].nunique()

total_products = df["Product ID"].nunique()

total_countries = df["Country"].nunique
```

```python
print("Total Sales: " , total_sales)
print("Total PRofit: ", total_profit)
print("Total Orders: ",total_orders)
print("Total Customer: ",total_customers)
print("Total Products: ",total_products)
print("Total Countries: " ,total_countries)
```

# =====================================
# 5. Category Analysis
# =====================================

```python
import matplotlib.pyplot as plt
category_sales=(
    df.groupby("Category")["Sales"].sum()
    .sort_values(ascending=False)
)
category_sales.plot(kind="bar"
                    ,figsize=(8,6)
                    )

plt.title("Sales By Category")
plt.xlabel("Category")
plt.ylabel("Sales")
plt.xticks(rotation=0)
plt.show()
```

```python
category_profit = (
    df.groupby("Category")["Profit"].sum()
    .sort_values(ascending=False)
)

category_profit.plot(
                    kind="bar",
                     figsize=(8,5)
                     )

plt.title("Profit By Category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.xticks(rotation=0)
plt.show()
```

# =====================================
# 6. Product Analysis
# =====================================

```python
top_products = (
    df.groupby("Product Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)

)

top_products.plot(
    kind="bar",
    figsize=(25,10)
)

plt.title("Top 10 Products")
plt.xlabel("Products")
plt.ylabel("Sales")
plt.xticks(rotation=0)
plt.show()
```

# =====================================
# 7. Customer Analysis
# =====================================

```python
top_customers_sales = (
    df.groupby("Customer Name")["Sales"].sum()
    .sort_values(ascending=False)
    .head(10)
)

top_customers_sales
```

```python
top_customers_profit = (
    df.groupby("Customer Name")["Profit"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_customers_profit
```

# =====================================
# 8. Geographic Analysis
# =====================================

```python
top_countries = (
    df.groupby("Country")["Sales"]
      .sum()
      .sort_values(ascending=False)
      .head(10)
)

top_countries.plot(
    kind="barh",
    figsize=(10,6)
)

plt.title("Top 10 Countries by Sales")
plt.xlabel("Sales")
plt.ylabel("Country")

plt.show()
```

# =====================================
# 9. Segment Analysis
# =====================================

```python
segment_analysis = (
    df.groupby("Segment")[["Sales","Profit"]].sum()
)

segment_analysis
```

# =====================================
# 10.Subcategory Analysis
# =====================================

```python
top_subcategories = (
    df.groupby("Sub-Category")["Sales"]
.sum()
.sort_values(ascending=False)

.head(10))

top_subcategories
```

```python
top_subcategories = (
    df.groupby("Sub-Category")["Sales"]
.sum()
.sort_values(ascending=False)

.head(10))

top_subcategories.plot(
    kind="bar",
    figsize=(10,6)
)

plt.title("Top 10 SubCategories by Sales")
plt.xlabel("Subcategories")
plt.ylabel("Sales")
plt.xticks(rotation=0)
plt.show()
```
